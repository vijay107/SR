Helper Function

```python
import math
```

Key Generation

```python
P = 59
Q = 61
```

```python
n = P * Q
phi_n = (P-1) * (Q-1)
```

```python
# Generating e
e = 2
while e < phi_n:
  if math.gcd(e, phi_n) == 1:
    break
  e += 1
```

```python
# Generating d
k = 1
while (k * phi_n + 1) % e != 0:
  k += 1
d = (k * phi_n + 1) // e  # // used for floor division
```

```python
U = [e, n] # Public key
R = [d, n] # Private key
```

```python
print("Primes:\t\t", P, ",", Q)
print("N:\t\t", n)
print("phi(N):\t\t", phi_n)
print("e:\t\t", e)
print("d:\t\t", d)
print("Public key:\t", "[e, n] =", U)
print("Private key:\t", "[d, n] =", R)
```

```python
plaintext = int(input("Enter Plaintext"))
```

```python
C= (plaintext**e) % n
print(C)
```

```python
P= (C**d) % n
print(P)
if P== plaintext:
 print("Message is succesfully shared")
else:
 print("Not shared")
```
