```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn import tree

#reading Dataset
dataset=pd.read_csv("data.csv")
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,5]
print("X",X)
print("Y",y)

#Perform Label encoding (Converting labels into numbers.)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

X=X.apply(le.fit_transform)


from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(X.iloc[:,1:5],y)  #ID not passed for training.

#Predict value for the given Expression
X_in=np.array([1,1,0,0]) # 1st 1--> Age<21, 2nd 1-->Income Low, 3rd 0--> Female, 4th-->Married
y_pred=clf.predict([X_in])
print("Prediction:", y_pred)

fig = plt.figure(figsize=(20,20))
_ = tree.plot_tree(clf, filled=True, feature_names=X.columns[[1,2,3,4]], label='all', rounded=True)
```
