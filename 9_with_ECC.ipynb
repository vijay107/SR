```python
class EllipticCurve:
  def __init__(self, a, b):
    self.a = a
    self.b = b
  
  def add(self, p1, p2, m = P):
    l = 0
    if p1 == p2:
      num = 3 * p1.x * p1.x + self.a
      den = 2 * p1.y
    else:
      num = p2.y - p1.y
      den = p2.x - p1.x
    l = moddiv(num, den, m)
    x3 = (l*l - p1.x - p2.x) % m
    y3 = (l*(p1.x - x3) - p1.y) % m
    return Point(x3, y3)

  def mul(self, k, p): # p is always generator point G
    temp = p
    while k != 1:
      temp = self.add(temp, p)
      k -= 1 
    return temp
  
  def sub(self, p1, p2):
    np = Point(p2.x, -p2.y)
    return self.add(p1, np)
```

```python
curve = EllipticCurve(2, 4) # Points lying on this curve:{0, 2}, {0, 5}, {1, 0}, {2, 3}, {2, 4}, {3, 3}, {3, 4}, {6, 1}, {6, 6}
G = Point(0, 2)
```

```python
def encrypt(p, U):
  k = 5
  c = [
       curve.mul(k, G),
       curve.add(p, curve.mul(k, U))
  ]
  return c
```

```python
def decrypt(C, R):
  p = curve.sub(C[1], curve.mul(R, C[0]))
  return p
```

```python
R = 5               # Private key
U = curve.mul(R, G) # Public key
```

```python
plaintext = Point(3, 4)
```

```python
ciphertext = encrypt(plaintext, U)
p = decrypt(ciphertext, R)
```
